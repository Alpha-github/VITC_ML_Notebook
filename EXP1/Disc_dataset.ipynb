```python
import random
import csv
```

```python
headers = ['index','Name','Roll no.','OS','DAA','ML','Gender']

num = 500
gen = ['M','F']
streams = ['BAI','BRS','CPS','CSE']
```

```python
records=[]

for i in range(num):
    t = []
    t.append(i+1)

    name = ''
    for j in range(10):
        name += chr(97 + random.randint(0,25))
    t.append(name)
        
    t.append('21'+random.choice(streams) + str(random.randint(1000,1600)))
    t.append(random.randint(0,101))
    t.append(random.randint(0,101))
    t.append(random.randint(0,101))
    t.append(random.choice(gen))
    records.append(t)

records[4]
```

```python
with open('Marks_data.csv','w',newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(headers)
    csvwriter.writerows(records)
```

Inference

```python
data = []
with open('Marks_data.csv','r') as f:
    csvreader = csv.reader(f)
    header = next(csvreader)
    for row in csvreader:
        data.append(row)
```

```python
for i in range(len(data)):
    rec = data[i]
    
    tot = int(rec[3]) + int(rec[4]) + int(rec[5])
    avg = tot/3

    if avg>40:
        status = 'P'
    else:
        status = 'F'
    
    if avg>=91 and avg<=100:
        grade = "S"
    elif avg>=81 and avg<91:
        grade = "A"
    elif avg>=71 and avg<81:
        grade = "B"
    elif avg>=61 and avg<71:
        grade = "C"
    elif avg>=51 and avg<61:
        grade = "D"
    elif avg>=41 and avg<51:
        grade = "E"
    else:
        grade = "F"
    
    rec.extend([tot,avg,grade,status])
```
